```python
import os
from matplotlib import pyplot as plt
import tensorflow as tf
import tensorflow_io as tfio

# Define file paths
CAPUCHIN_FILE = os.path.join('/content/drive/MyDrive/data (1)', 'Parsed_Capuchinbird_Clips', 'XC3776-3.wav')
NOT_CAPUCHIN_FILE = os.path.join('/content/drive/MyDrive/data (1)', 'Parsed_Not_Capuchinbird_Clips', 'afternoon-birds-song-in-forest-0.wav')

# Function to load and resample WAV files
def load_wav_16k_mono(filename):
    file_contents = tf.io.read_file(filename)
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    return wav

# Load example audio files
wave = load_wav_16k_mono(CAPUCHIN_FILE)
nwave = load_wav_16k_mono(NOT_CAPUCHIN_FILE)

# Plot the waveforms
plt.plot(wave)
plt.plot(nwave)
plt.show()

# Define directories for positive and negative examples
POS = os.path.join('/content/drive/MyDrive/data (1)', 'Parsed_Capuchinbird_Clips')
NEG = os.path.join('/content/drive/MyDrive/data (1)', 'Parsed_Not_Capuchinbird_Clips')

# Create datasets
pos_files = tf.data.Dataset.list_files(POS + '/*.wav')
neg_files = tf.data.Dataset.list_files(NEG + '/*.wav')

# Label positive and negative datasets
positives = pos_files.map(lambda x: (x, 1))
negatives = neg_files.map(lambda x: (x, 0))

# Combine positive and negative datasets
data = positives.concatenate(negatives)

# Preprocess the data: load, pad/truncate, and generate spectrogram
def preprocess(file_path, label):
    def load_wav(filename):
        filename = filename.numpy().decode('utf-8')  # Convert TensorFlow tensor to string
        wav = load_wav_16k_mono(filename)
        wav = wav[:48000]  # Truncate or pad audio to 48000 samples
        zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
        wav = tf.concat([zero_padding, wav], 0)
        return wav

    # Load audio using tf.py_function to allow Python file handling
    wav = tf.py_function(load_wav, [file_path], tf.float32)
    wav.set_shape([48000])  # Set shape explicitly to avoid dynamic shape issues

    # Generate spectrogram
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)

    # Ensure fixed spectrogram size
    spectrogram = spectrogram[:1491, :257]  # Truncate or pad spectrogram to (1491, 257)
    spectrogram = tf.expand_dims(spectrogram, axis=2)  # Add channel dimension

    return spectrogram, label

# Apply preprocessing to the dataset
data = data.map(preprocess)
data = data.cache()
data = data.shuffle(buffer_size=1000)
data = data.batch(16)
data = data.prefetch(8)

# Split into training and testing datasets
train = data.take(36)
test = data.skip(36).take(15)

# Debug: Check the shape of the dataset inputs and labels
for batch in train.take(1):
    inputs, labels = batch
    print(f"Input shape: {inputs.shape}")
    print(f"Label shape: {labels.shape}")

# Define the CNN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, GlobalAveragePooling2D

model = Sequential()
model.add(Conv2D(16, (3, 3), activation='relu', input_shape=(1491, 257, 1)))
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(GlobalAveragePooling2D())  # Replace Flatten with Global Pooling
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='Adam',
              loss='BinaryCrossentropy',
              metrics=[tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])

# Print the model summary
model.summary()

# Train the model
hist = model.fit(train, epochs=4, validation_data=test)
model.save('capuchin_model.h5')

# plt.title('Loss')
# plt.plot(hist.history['loss'], 'r')
# plt.plot(hist.history['val_loss'], 'b')
# plt.show()

# plt.title('Precision')
# plt.plot(hist.history['precision'], 'r')
# plt.plot(hist.history['val_precision'], 'b')
# plt.show()

# plt.title('Recall')
# plt.plot(hist.history['recall'], 'r')
# plt.plot(hist.history['val_recall'], 'b')
# plt.show()


X_test, y_test = test.as_numpy_iterator().next()

yhat = model.predict(X_test)

yhat = [1 if prediction > 0.5 else 0 for prediction in yhat]

def load_mp3_16k_mono(filename):
    """ Load a WAV file, convert it to a float tensor, resample to 16 kHz single-channel audio. """
    res = tfio.audio.AudioIOTensor(filename)
    # Convert to tensor and combine channels
    tensor = res.to_tensor()
    tensor = tf.math.reduce_sum(tensor, axis=1) / 2
    # Extract sample rate and cast
    sample_rate = res.rate
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    # Resample to 16 kHz
    wav = tfio.audio.resample(tensor, rate_in=sample_rate, rate_out=16000)
    return wav

mp3 = os.path.join('/content/drive/MyDrive/data (1)/', 'Forest Recordings', 'recording_00.mp3')
wav = load_mp3_16k_mono(mp3)

audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)

samples, index = audio_slices.as_numpy_iterator().next()


def preprocess_mp3(sample, index):
    sample = sample[0]
    zero_padding = tf.zeros([48000] - tf.shape(sample), dtype=tf.float32)
    wav = tf.concat([zero_padding, sample],0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram



audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=16000, sequence_stride=16000, batch_size=1)
audio_slices = audio_slices.map(preprocess_mp3)
audio_slices = audio_slices.batch(64)

yhat = model.predict(audio_slices)
yhat = [1 if prediction > 0.5 else 0 for prediction in yhat]

from itertools import groupby

yhat = [key for key, group in groupby(yhat)]
calls = tf.math.reduce_sum(yhat).numpy()

calls

results = {}
for file in os.listdir(os.path.join('/content/drive/MyDrive/data (1)/', 'Forest Recordings')):
    FILEPATH = os.path.join('/content/drive/MyDrive/data (1)/','Forest Recordings', file)

    wav = load_mp3_16k_mono(FILEPATH)
    audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)
    audio_slices = audio_slices.map(preprocess_mp3)
    audio_slices = audio_slices.batch(64)

    yhat = model.predict(audio_slices)

    results[file] = yhat

results

class_preds = {}
for file, logits in results.items():
    class_preds[file] = [1 if prediction > 0.99 else 0 for prediction in logits]
class_preds

postprocessed = {}
for file, scores in class_preds.items():
    postprocessed[file] = tf.math.reduce_sum([key for key, group in groupby(scores)]).numpy()
postprocessed

import csv

with open('results.csv', 'w', newline='') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['recording', 'capuchin_calls'])
    for key, value in postprocessed.items():
        writer.writerow([key, value])
```

```python
import os
import csv
import numpy as np
import ipywidgets as widgets
from IPython.display import display, Audio
import tensorflow as tf
import tensorflow_io as tfio
from matplotlib import pyplot as plt

# Define a function to load a WAV file and resample it to 16kHz mono
def load_wav_16k_mono(filename):
    file_contents = tf.io.read_file(filename)
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    return wav

# Function to load and preprocess an MP3 file
def load_mp3_16k_mono(filename):
    res = tfio.audio.AudioIOTensor(filename)
    tensor = res.to_tensor()
    tensor = tf.math.reduce_sum(tensor, axis=1) / 2
    sample_rate = res.rate
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    wav = tfio.audio.resample(tensor, rate_in=sample_rate, rate_out=16000)
    return wav

# Preprocess the MP3 file to extract spectrogram
def preprocess_mp3(sample, index):
    sample = sample[0]
    zero_padding = tf.zeros([48000] - tf.shape(sample), dtype=tf.float32)
    wav = tf.concat([zero_padding, sample], 0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram

# Function to run inference on uploaded audio files
def run_inference(file_path):
    wav = load_mp3_16k_mono(file_path)
    audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)
    audio_slices = audio_slices.map(preprocess_mp3)
    audio_slices = audio_slices.batch(64)

    # Load model and predict
    yhat = model.predict(audio_slices)
    predictions = [1 if prediction > 0.5 else 0 for prediction in yhat]

    return predictions

# Display predictions in the frontend
def display_predictions(predictions, file_name):
    capuchin_calls = sum(predictions)
    print(f"File: {file_name}, Capuchinbird Calls: {capuchin_calls}")

# Load the trained model
model = tf.keras.models.load_model('capuchin_model.h5')

# File upload widget
file_upload = widgets.FileUpload(accept='.mp3', multiple=False)

# Button to start inference
button = widgets.Button(description="Run Inference")

# Output widget to display the results
output = widgets.Output()

# Function to process uploaded file
def on_button_clicked(b):
    with output:
        output.clear_output()
        if file_upload.value:
            file_info = list(file_upload.value.values())[0]
            file_name = file_info['metadata']['name']
            content = file_info['content']
            with open(file_name, 'wb') as f:
                f.write(content)

            predictions = run_inference(file_name)
            display_predictions(predictions, file_name)
            display(Audio(file_name))
        else:
            print("Please upload a file to process.")

# Connect the button click event
button.on_click(on_button_clicked)

# Display the widgets
display(widgets.VBox([file_upload, button, output]))
```
